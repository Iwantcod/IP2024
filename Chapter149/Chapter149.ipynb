```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as pl
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./20241028_122920.png',0)
edges100 = cv2.Canny(img,100,100)
edges200 = cv2.Canny(img,200,200)
edges = cv2.Canny(img,100,200)
```

```python
contours, hierachy = cv2.findContours(edges,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
img0 = img.copy()
img1 = cv2.drawContours(img0, contours, 0, (255,0,0),5)
```

```python
plt.figure(figsize=(24,8))
plt.subplot(1,2,1),plt.imshow(edges)
plt.title('original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(img1)
plt.title('Coutour'), plt.xticks([]), plt.yticks([])
plt.show()
```
