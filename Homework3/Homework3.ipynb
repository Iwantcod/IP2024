```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dsu3.jpg')
rows,cols,ch = img.shape

# 역 디귿자 순서로 좌표를 추출
# output은 해당 이미지만을 추출하므로 시작 좌표가 0,0이다.(output은 사실상 비율 조정하는 란)
pts1 = np.float32([[106,226],[703,163],[705,539],[109,497]])
pts2 = np.float32([[0,0],[1500,0],[1500,800],[0,800]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(1500,800))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
